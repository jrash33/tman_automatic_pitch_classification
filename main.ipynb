# Missing Values

```python
#dealing with NaN values

#1 find how many NaN values in each row
NaN_series = df.isnull().sum(axis=1).sort_values(ascending=False)

#find a good threshold of rows to just delete completely
#final: delete all rows with more than 5 NaN values (coverage too poor)
thresh = 5
bad_rows = NaN_series[(NaN_series > 0) == True].shape
worst_rows = NaN_series[(NaN_series > thresh) == True].shape
print(f"Number of Rows w NaN Values: {bad_rows[0]}")
print(f"Number of Rows w more than {thresh} NaN Values: {worst_rows[0]}")
```

```python
#input medians for NaNs (all spinRate) grouped by pitchTypeCode and pitcherId
#This is because we don't have much data and can't afford to delete too many rows - must approximate unless bigger data set
rows_to_fix = NaN_series[(NaN_series > 0) == True].index

for i in rows_to_fix:
    pitcher = df.iloc[i]['pitcherId']
    pitch = df.iloc[i]['pitchTypeCode']
    x = df.loc[(df['pitchTypeCode'] == pitch) & (df['pitcherId'] == pitcher)]['spinRate'].median()
    df.at[i, 'spinRate'] = x
```

```python
#now delete rows that have NaN values above our set threshold of 5 NaN values
data_drop = NaN_series[(NaN_series > thresh) == True].index
df = df.drop(data_drop)
print(df.shape)
```

```python
#Delete columns we know we don't need in order to build our model (by intuition/observation)
df = df.iloc[:,1:]
df.drop(['gamePk', 'inning', 'halfInning', 'atBatIndex', 'pitchIndex', 'playId'],axis=1,inplace=True)
print(df.shape)
df.head()
```

# Feature Engineering

### create interaction variables

```python
#first before anything - split the data

#train_data, test_data = train_test_split(df, test_size=.2)
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,3:], df['pitchTypeCode'], 
                                                    stratify=Y, test_size=0.2, random_state = 100)

print(X_train.shape) 
print(X_test.shape)

#due to random sampling, we need to reset our index values
X_train.reset_index(inplace=True, drop=True)
X_test.reset_index(inplace=True, drop=True)
y_train.reset_index(inplace=True, drop=True)
y_test.reset_index(inplace=True, drop=True)
```

```python
#transform data set by adding interaction terms of degree 2
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = pd.DataFrame(poly_features.fit_transform(X_train))
#poly_coef = poly_features.get_feature_names()

#display interaction variable for column name
X_poly.columns = poly_features.get_feature_names(X_train.columns)
X_poly.head()
```

### Scale the data

```python
min_max_scaler = preprocessing.MinMaxScaler() #we want a scale from 0 to 1 with no negatives and keeping centers the same
X_poly_scaled = pd.DataFrame(min_max_scaler.fit_transform(X_poly))

#columns
X_poly_scaled.columns = list(X_poly)

X_poly_scaled.head()
```

### Feature Selection

```python
#a couple methods we could use to reduce number of features to prevent noise and overfitting- let's see what each one gives us
# 1.) basic variance test
# 2.) F-test/chi-square test
# 3.) RFE (random forest)
# 4.) Tree based (select from model method)

#Let's use RFE because we have time
```

```python
#first - remove features with low variance

threshold = .5 #standard
sel_variance_threshold = VarianceThreshold(threshold) 
#fit variance threshold to X_poly data
sel_variance_threshold.fit_transform(X_poly)

#delete columns with low variance
X_poly_scaled = pd.DataFrame(X_poly_scaled[X_poly_scaled.columns[sel_variance_threshold.get_support(indices=True)]])
print(f"new data shape: {X_poly_scaled.shape}")
```

```python
#RFE(Recursive Feature Elimination) AKA Backward Selection (train -> eliminate least important param -> retrain)
model_tree = RandomForestClassifier(random_state=100, class_weight='balanced') #class_weight param to balance classes
sel_rfe_tree = RFE(estimator=model_tree, n_features_to_select=30, step=1) #n_feature num can be optimized in a later model
X_train_rfe_tree = sel_rfe_tree.fit_transform(X_poly_scaled, y_train)
print(sel_rfe_tree.get_support())
```

we can see the 30 columns optimized for importance

```python
var = sel_rfe_tree.get_support()
good_var = np.where(var)[0]
good_var #these are the column numbers of 30 most important features
```

```python
#now grab these 30 important feature columns
X_train_opt = X_poly_scaled.iloc[:,good_var]

#here is our final data set cleaned and optimized for modeling!
X_train_opt.head()
```

Final Data set to be filtered into model

### RandomForest with hyperparameters tuned

```python
#HYPERPARAMETER TUNING
from sklearn.model_selection import RandomizedSearchCV
from pprint import pprint

rf = RandomForestClassifier(random_state = 42, class_weight='balanced')

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
#class weight
class_weight = ['balanced', 'balanced_subsample', None]
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 500, num = 10)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'class_weight': class_weight,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train_opt, y_train)
```

### CONCLUSION:
The KNN model behaves the most accurately and effectively for this multi-classifation problem. The reasons for this are 2 fold because this data is imbalanced in 2 different ways. There is a large imbalance of pitch types and pitches thrown by pitchers. Because of this, I am looking primarily at my classification/confusion matrix while evaluating my model opposed to accuracy.

Because I see a better f1-score for my KNN model (higher accuracy as well), this is my model of choice which I conclude to have a 95% accuracy

### Ways to improve model

Looking at the confusion matrix/classification report, we can see that the error in large part is due to cutter/curveball misclassifications as sliders. Digging deeper into each model, it is seen that every model has a difficult time distinguishing cutters and sliders.

-> If I had more time, I would take these steps to improve my model:
1. I used a class weighting hyperparameter to account for the class imbalance. Another method would be to try over/under sampling and see how the model responds. I could even oversample the cutters/curveballs more to give the model more efficiency with these pitches
2. Dig deeper into cutters/curveballs and reasons as to why the model would mix them up with sliders
3. Optimize my feature reduction - I set a hard limit of 30 features. If I had more time, I could optimize the number of features to decrease possible overfitting. This optimal number could be much more or much less.
4. I was brainstorming ways to account for the pitcherID imbalance if this became an issue (ended up being ok but could potentially account for some error). One thought is to bin players by speed and build a couple different models depending on player type/parameters. I could also try over/under sampling by pitcherID and see how the model reacts

Overall, I am pleased with the results but also feel this model can be improved upon in terms of the FC/CU/SL classifications
